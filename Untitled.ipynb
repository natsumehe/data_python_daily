```python
import sqlite3
import pandas as pd

# 连接到 Case Study 1 数据库
conn_case_study_1 = sqlite3.connect('database1.db')

# 连接到 Case Study 2 数据库
conn_case_study_2 = sqlite3.connect('database2.db')

# 连接到 Case Study 4 数据库
conn_case_study_4 = sqlite3.connect('database4.db')

# 从 Case Study 1 加载数据
query_case_study_1 = "SELECT * FROM data_table"
df_case_study_1 = pd.read_sql(query_case_study_1, conn_case_study_1)

# 从 Case Study 2 加载数据
query_case_study_2 = "SELECT * FROM data_table2"
df_case_study_2 = pd.read_sql(query_case_study_2, conn_case_study_2)

# 从 Case Study 4 加载数据
query_case_study_4 = "SELECT * FROM data_case4"
df_case_study_4 = pd.read_sql(query_case_study_4, conn_case_study_4)

# 合并 Case Study 1 和 Case Study 2
df_merged_1_2 = pd.merge(df_case_study_1, df_case_study_2, on='tr_uniq_id', how='inner')

# 合并上一步的结果与 Case Study 4
df_merged = pd.merge(df_merged_1_2, df_case_study_4, on='tr_uniq_id', how='inner')

# 筛选出 Card Country Code 与 Customer Country 不一致的订单
df_mismatch = df_merged[df_merged['card_country_code'] != df_merged['customer_country']]

df_mismatch = df_mismatch.copy()

# Now mark whether it's a fraud order
df_mismatch.loc[:, 'is_fraud'] = (df_mismatch['op_error_code'] == 'card.possible-fraud')

# 计算总体欺诈率
total_orders = df_mismatch.shape[0]
fraud_orders = df_mismatch['is_fraud'].sum()
overall_fraud_rate = (fraud_orders / total_orders) * 100

# 输出总体欺诈率
print(f"总体欺诈率：{overall_fraud_rate:.2f}%")

# 计算每个 Card IIN 的欺诈率
fraud_rates_by_iin = df_mismatch.groupby('card_iin').agg(
    total_orders=('is_fraud', 'size'),
    fraud_orders=('is_fraud', 'sum')
).reset_index()

# 计算每个 Card IIN 的欺诈率
fraud_rates_by_iin['fraud_rate'] = fraud_rates_by_iin['fraud_orders'] / fraud_rates_by_iin['total_orders'] * 100

# 排序并显示前五个 Card IIN 的欺诈率
top_fraud_by_iin = fraud_rates_by_iin.sort_values(by='fraud_rate', ascending=False).head(5)

# 显示结果
print(top_fraud_by_iin[['card_iin', 'fraud_rate']])

# 关闭数据库连接
conn_case_study_1.close()
conn_case_study_2.close()
conn_case_study_4.close()
```

```python
import sqlite3
import pandas as pd
import re

# 连接到 Case Study 1 数据库
conn_case_study_1 = sqlite3.connect('database1.db')

# 连接到 Case Study 2 数据库
conn_case_study_2 = sqlite3.connect('database2.db')

# 连接到 Case Study 4 数据库
conn_case_study_4 = sqlite3.connect('database4.db')

# 从 Case Study 1 加载数据
query_case_study_1 = "SELECT * FROM data_table"
df_case_study_1 = pd.read_sql(query_case_study_1, conn_case_study_1)

# 从 Case Study 2 加载数据
query_case_study_2 = "SELECT * FROM data_table2"
df_case_study_2 = pd.read_sql(query_case_study_2, conn_case_study_2)

# 从 Case Study 4 加载数据
query_case_study_4 = "SELECT * FROM data_case4"
df_case_study_4 = pd.read_sql(query_case_study_4, conn_case_study_4)

# 合并 Case Study 1 和 Case Study 2
df_merged_1_2 = pd.merge(df_case_study_1, df_case_study_2, on='tr_uniq_id', how='inner')

# 合并上一步的结果与 Case Study 4
df_merged = pd.merge(df_merged_1_2, df_case_study_4, on='tr_uniq_id', how='inner')

# 提取 customer_email 中的域名部分
df_merged['email_domain'] = df_merged['customer_email'].str.extract(r'@([A-Za-z0-9.-]+)')

# 标记是否为欺诈订单
df_merged['is_fraud'] = df_merged['op_error_code'] == 'card.possible-fraud'

# 计算每个 email_domain 的欺诈率
fraud_rates_by_domain = df_merged.groupby('email_domain').agg(
    total_orders=('is_fraud', 'size'),
    fraud_orders=('is_fraud', 'sum')
).reset_index()

# 计算欺诈率
fraud_rates_by_domain['fraud_rate'] = fraud_rates_by_domain['fraud_orders'] / fraud_rates_by_domain['total_orders'] * 100

# 按欺诈率降序排序
sorted_fraud_rates = fraud_rates_by_domain.sort_values(by='fraud_rate', ascending=False)

# 显示结果
print("电子邮箱域名的欺诈率：")
print(sorted_fraud_rates[['email_domain', 'fraud_rate']])

# 筛选出欺诈订单
fraud_orders_by_domain = df_merged[df_merged['is_fraud']]

# 按 email_domain 分组并列出 Card IIN 和 CustomerCountry
fraud_orders_by_domain_info = fraud_orders_by_domain.groupby('email_domain').agg(
    card_iins=('card_iin', 'unique'),
    customer_countries=('customer_country', 'unique')
).reset_index()

# 显示每个域名的欺诈订单所使用的 Card IIN 和 Customer Country
print("\n每个域名的欺诈订单所使用的 Card IIN 和 Customer Country：")
print(fraud_orders_by_domain_info[['email_domain', 'card_iins', 'customer_countries']])

# 关闭数据库连接
conn_case_study_1.close()
conn_case_study_2.close()
conn_case_study_4.close()
```

```python
import sqlite3
import pandas as pd

# 连接到 Case Study 1 数据库
conn_case_study_1 = sqlite3.connect('database1.db')

# 连接到 Case Study 2 数据库
conn_case_study_2 = sqlite3.connect('database2.db')

# 连接到 Case Study 4 数据库
conn_case_study_4 = sqlite3.connect('database4.db')

# 从 Case Study 1 加载数据
query_case_study_1 = "SELECT * FROM data_table"
df_case_study_1 = pd.read_sql(query_case_study_1, conn_case_study_1)

# 从 Case Study 2 加载数据
query_case_study_2 = "SELECT * FROM data_table2"
df_case_study_2 = pd.read_sql(query_case_study_2, conn_case_study_2)

# 从 Case Study 4 加载数据
query_case_study_4 = "SELECT * FROM data_case4"
df_case_study_4 = pd.read_sql(query_case_study_4, conn_case_study_4)

# 合并 Case Study 1 和 Case Study 2
df_merged_1_2 = pd.merge(df_case_study_1, df_case_study_2, on='tr_uniq_id', how='inner')

# 合并上一步的结果与 Case Study 4
df_merged = pd.merge(df_merged_1_2, df_case_study_4, on='tr_uniq_id', how='inner')

# 筛选出 3DS 订单
df_3ds = df_merged[df_merged['op_type'] == 'three_d_s_2_check']

# 计算每个 CustomerCountry 中成功和失败订单的比例
df_3ds_grouped = df_3ds.groupby('customer_country').agg(
    total_orders=('op_has_failed', 'size'),
    successful_orders=('op_has_failed', lambda x: (x == False).sum()),
    failed_orders=('op_has_failed', lambda x: (x == True).sum())
).reset_index()

# 计算成功和失败的比例
df_3ds_grouped['success_rate'] = df_3ds_grouped['successful_orders'] / df_3ds_grouped['total_orders'] * 100
df_3ds_grouped['failure_rate'] = df_3ds_grouped['failed_orders'] / df_3ds_grouped['total_orders'] * 100

# 排序并显示结果
df_3ds_grouped_sorted = df_3ds_grouped.sort_values(by='success_rate', ascending=False)

print("每个 CustomerCountry 中 3DS 订单的成功率和失败率：")
print(df_3ds_grouped_sorted[['customer_country', 'success_rate', 'failure_rate']])

# 关闭数据库连接
conn_case_study_1.close()
conn_case_study_2.close()
conn_case_study_4.close()
```
